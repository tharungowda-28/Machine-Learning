Perform Exploratory Data Analysis on England tour of India 2021 Dataset

Q 1: Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Q2 : Load the 3 datasets ( batting, bowling and combined )

```python
bat = pd.read_csv('/content/batting.csv')
bowl = pd.read_csv('/content/bowling.csv')
comb = pd.read_csv('/content/combined.csv')
```

Q 3: Display top 3 values of batting DataFrame

```python
bat.head(3)
```

Q4 : Display no of rows and columns in every dataframe

```python
print(bat.shape)
print('There are 31 rows and 14 columns in Batting CSV file')
print(bowl.shape)
print('There are 16 rows and 13 columns in Bowling CSV file')
print(comb.shape)
print('There are 31 rows and 12 columns in Combined CSV file')
```

Q5: Top 5 run scorers

```python
top5 = comb.sort_values(by=['runs_scored'],ascending=False).head()
top5.iloc[:,0:3]
```

Q6 : Plot a graph for top 5 run scorers in the series

```python
plt.plot(top5['Player'],top5['runs_scored'])
plt.xlabel('Players')
plt.ylabel('Runs Scored')
plt.show()
```

Q7: Top 5 wicket takers

```python
bowl5 = bowl.sort_values(by=['wickets'],ascending=False).head()
bowl5.loc[:,['Player','wickets']]
```

Q8 : Plot the top 5 wicket takers of the series ( Bar Graph)

```python
plt.bar(bowl5['Player'],bowl5['wickets'])
plt.xlabel('Players')
plt.ylabel('Wickets')
plt.show()
```

Q9: Best All-round performers

```python
batavg = comb[comb['bat_avg'].values>20]
x = batavg[batavg['bw_avg'].values<30]
y = x[x['bw_avg'].values>20]
#bowlavg = bowlavg.sort_values(by=['bw_avg'])
#batavg = batavg.sort_values(by=['bat_avg'],ascending=False)
#bl = bowlavg['bw_avg'].min()
#y = bowlavg[bowlavg['bat_avg'].values>bowlavg['bw_avg'].values]
y.loc[:,['Player']]
```

Q 10: Plot the best allrounder graph ( Bar Graph)

```python
plt.bar(y['Player'],y['bat_avg'],color='red',label='Batting Average',width=0.4,align='edge')
plt.bar(y['Player'],y['bw_avg'],color='yellow',label='Bowling Average',width=-0.4,align='edge')
plt.legend()
plt.show()
```
